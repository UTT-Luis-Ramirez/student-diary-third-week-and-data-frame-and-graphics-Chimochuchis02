###**Anotaciones Semana 2 prob y estat**###


***Poblacion***: Es el conjunto de individuos que vamos a utilizar para nuestros analisis.

***P.Finita:*** **Tiene un fin aparente**

***P.Infinita:*** **no tiene fin**

***Muestras:***Es el conjunto o subconjunto representativo que es seleccionado para la población, que incluya todas las caracteristicas de la población.

***1-*** Debe ser **REPRESENTATIVA**

***2-***Tiene que tener un tamaño **DEMOSTRATIVO**

***3-***Tiene que ser **DIVERSO** los **DATOS**

***Paramétro:*** Valor numerico que describe una característica de la población.

***Valor estadístico:***Es lo **CONTRARIO** del **PARAMETRO**, es usado para asumir o determinar algo de la población.

***Muestreo:***Este tiene 2 tipos.

***Aleatorio o probabilistico:***tiene una igual probabilidad para la muestra.

***No probabilistico-SISTEMATICO:*** Un sistema aleatorio para elegir.

***Aleatorio simple:***Numeros al azar con programacion.

***Datos estartificados:***Dividir nuestro universo en conjuntos llamados estratos, seleccionados proporcionalmente.

***Por Conglomerados:*** Se seleccionan los grupos para haacer una muestra de una poblacion grande.

***Muestreo no probabilistica:*** muestras no aleatorias.

segmentacion por **CONVENIENCIA**

**MUESTREO** por **JUICIO**

***Muestreo por cuotas:***con proporciones especificas.

###Obtener el tamañao de la poblacion

##creacion de una muestra aleatoria simple de 50 registros

```python
sample_size = 50
sample = data.sample(n= sample_size, random_state=12)
```

```python
population_size = len(data)
print("el tamaño de la poblacion es",population_size)
```

##Revision de la Muestra

```python
display(sample)
```

```python
sample_size = len(sample)
print(sample_size)
```

##Muestreo Aleatorio Sistematico

##seleccion sistematica de la mauestra

##Definicion del intervalo k(tamaño de la poblacion / el tamaño de la muestra deseada)

```python
sample_size = 100
interval_k = population_size//sample_size
```

##seleccion de muestra sistematica

```python
systematic_sample = data.iloc[::interval_k]
```

```python
print("Muestra obtenida por muestreo sistematico")
display(systematic_sample)
```

### Agrupacion de datos y subpoblaciones

## dividir la poblacion por un factor

```python
grouped_by_dev_language = data.groupby("Idioma de programación")
grouped_by_dev_language.head()
```

```python
group_analysis = grouped_by_dev_language[["Líneas de código","Número de bugs"]].mean()
print("analisis por Tamaño de equipo")
display(group_analysis)
```

##Analisis por tiempo y numero del equipo.


```python
group_analysis = grouped_by_dev_language[["Tiempo de desarrollo (horas)","Tamaño del equipo"]].mean()
print("analisis por lenguaje de programacion")
display(group_analysis)
```
